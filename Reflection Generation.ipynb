```python
from google.colab import files
from io import BytesIO
from PIL import Image

uploaded = files.upload()
image1 = Image.open(BytesIO(uploaded['images.jpg']))

```

```python
uploaded = files.upload()
image2 = Image.open(BytesIO(uploaded['images1.jpg']))
```

```python
from PIL import Image,ImageFilter,ImageEnhance
def changesize(maxWidth,maxHeight,image):
    widthRatio = maxWidth / image.size[0]
    heightRatio = maxHeight / image.size[1]
    newWidth = int(widthRatio * image.size[0])
    newHeight = int(heightRatio * image.size[1])
    newImage = image.resize((newWidth, newHeight))
    return newImage
image3 = changesize(612, 400, image1)
image4 = changesize(612, 400, image2)

```

```python
image4 = image4.filter(ImageFilter.GaussianBlur(radius = 2))
#image3.show()
#cv2.imshow('Gaussian Blurring', Gaussian)
#cv2.waitKey(0)

image5 = image3.convert("RGBA")
image6 = image4.convert("RGBA")
image5.show()
image6.show()
blend = Image.blend(image5, image6, alpha=0.200)
blend
```
